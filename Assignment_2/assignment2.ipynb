```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return 5 * x**2 + 2

def df(x):
    return 10 * x

# Gradient Descent Update Rule
def gradient_descent_update(x, eta):
    return x - eta * df(x)

# Visualizing Convergence
def plot_convergence(eta_values, num_iterations=20):
    x_values = np.linspace(-5, 5, 100)
    y_values = f(x_values)

    plt.figure(figsize=(12, 8))
    plt.plot(x_values, y_values, label='f(x) = 5x^2 + 2', linewidth=2)

    for eta in eta_values:
        x = 4.5  # Initial starting point
        x_iterates = [x]

        for _ in range(num_iterations):
            x = gradient_descent_update(x, eta)
            x_iterates.append(x)

        plt.scatter(x_iterates, [f(x) for x in x_iterates], label=f'η = {eta}', marker='o', alpha=0.5)

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gradient Descent Convergence with Different Learning Rates')
    plt.legend()
    plt.grid(True)
    plt.show()

# Test with different learning rates, including eta > 0.2
eta_values_extended = [0.01, 0.05, 0.1, 0.2, 0.3, 0.5]
plot_convergence(eta_values_extended, num_iterations=20)
```
